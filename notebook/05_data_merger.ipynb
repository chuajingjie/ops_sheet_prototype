```python
import pandas as pd
```

```python
knowledge_base_df = pd.read_csv("../data/sample/processed/knowledge_base.csv")
xy_mapping_df = pd.read_csv("../data/xy_mapping/processed/xy_mapping_20250414.csv")
```

```python
# Perform left join on knowledge_base_df with xy_mapping_df using Int. No and Side ID
merged_df = knowledge_base_df.merge(
    xy_mapping_df,
    how='left',
    left_on='Int. No',
    right_on='Site_ID'
)

merged_df.head(3)
```

```python
merged_df = merged_df.drop(columns=["Site_ID"])

merged_df.to_csv("../data/sample/processed/knowledge_base.csv", index=False)
```
