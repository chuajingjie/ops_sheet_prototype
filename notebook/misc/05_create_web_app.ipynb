```python
import streamlit as st
import pandas as pd
import requests

# Load your knowledge base CSV
csv_path = r"C:\Users\chuaj\OneDrive\Desktop\Folder\5. Work\10. Knowledge Base\data\sample\processed\knowledge_base.csv"
df = pd.read_csv(csv_path)
context = df.to_markdown(index=False)

# Streamlit UI
st.title("LLM Q&A Chatbot (LLaMA 3 Local via Ollama)")

question = st.text_input("Enter your question:")

if st.button("Ask"):
    if question.strip() == "":
        st.warning("Please enter a question.")
    else:
        prompt = f"""Refer to the following knowledge base and answer the question.

Knowledge Base (in table format):
{context}

Question: {question}

Output: Your answers should only give the intersection number (e.g. 106)
"""

        # Make the API call to Ollama
        with st.spinner("Thinking..."):
            response = requests.post(
                'http://localhost:11434/api/generate',
                json={
                    'model': 'llama3.1:8b',
                    'prompt': prompt,
                    'stream': False    
                }
            )

            result = response.json().get('response', '[No response]')
            st.success("Answer:")
            st.write(result)
```
